```python
# Load the "diamonds" dataset
diamonds_df = spark.read.format("csv").option("header", "true").load(f"{DATA_DIR}/databricks-datasets/Rdatasets/csv/ggplot2/diamonds.csv")
```

```python
diamonds_df.show()
```

```python
diamonds_df.count()
```

```python
filtered_df = diamonds_df.filter((diamonds_df["cut"] == "Ideal") & (diamonds_df["price"] > 1000))
```

```python
filtered_df.count()
```

```python
diamonds_df.count()
```

```python
filtered_df.show()
```

```python
from pyspark.sql.functions import avg


grouped_df = filtered_df.groupBy("clarity").agg(avg("price"))
```

```python
grouped_df.show()
```

```python
ordered_df = grouped_df.orderBy(grouped_df["avg(price)"].desc())
```

```python
ordered_df.show()
```

```python
ordered_df.show()
```

```python
ordered_df_new = ordered_df.withColumnRenamed("avg(price)", "avg_price")

```

```python
ordered_df_new.show()
```

```python
ordered_df_new.write.format("delta").mode("overwrite").saveAsTable("clarity_fx_skills")
```
